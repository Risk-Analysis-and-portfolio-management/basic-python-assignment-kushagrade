```python
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd

def fetch(ticker, start, end,interval):
    data = yf.download(tickers=ticker, start=start, end=end, interval=interval)
    return data
data=fetch ('AAPL',start="2023-12-13",end="2024-12-13",interval="1d")['Close']
def plot_stocks (Stock , name ):
    plt.figure(figsize=(10,5))
    plt.plot( Stock)
    plt.title(name )
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.show()
plot_stocks( data,'Apple Stock Price')
```

```python
def analyze_portfolio( ticker,weights,start_date,end_date):
    pf=fetch ( ticker, start =start_date,end =end_date,interval='1d')['Adj Close']
    num_aapl= (100000*weights[0] )/ pf['AAPL'].iloc[0]
    new_aapl = num_aapl.astype(int) *pf['AAPL']
    num_msft= (100000*weights[1]) / pf['MSFT'].iloc[0]
    new_msft = num_msft.astype(int) * pf['MSFT']
    num_goog= (100000*weights[2] )/ pf['GOOG'].iloc[0]
    new_goog = num_goog.astype(int) *pf['GOOG']
    num_tsla=( 100000*weights[3] )/ pf['TSLA'].iloc[0]
    new_tsla = num_tsla.astype(int) * pf['TSLA']
    num_nvda= (100000*weights[4] )/ pf['NVDA'].iloc[0]
    new_nvda = num_nvda.astype(int) * pf['NVDA'] 
    total = pd.DataFrame()
    total=new_goog + new_msft + new_aapl +new_nvda+new_tsla
    print("Initial Portfolio Value")
    print(total.iloc[0])
    print('Final Portfolio Value:')
    print(total.iloc[-1])
    plot_stocks(total, 'Portfolio')
analyze_portfolio(
    ticker=['AAPL', 'MSFT', 'GOOG', 'TSLA', 'NVDA'], 
    weights=[0.3, 0.2, 0.2, 0.2, 0.1], 
    start_date="2023-01-01", 
    end_date="2023-12-31"
)
    
    
```

```python
def analyze_better_portfolio( ticker,weights,start_date,end_date):
    pf=fetch ( ticker, start =start_date,end =end_date,interval='1d')['Adj Close']
    num_aapl= (100000*weights[0] )/ pf['AAPL'].iloc[0]
    new_aapl = num_aapl.astype(int) *pf['AAPL']
    num_msft= (100000*weights[1]) / pf['MSFT'].iloc[0]
    new_msft = num_msft.astype(int) * pf['MSFT']
    num_goog= (100000*weights[2] )/ pf['GOOG'].iloc[0]
    new_goog = num_goog.astype(int) *pf['GOOG']
    num_tsla=( 100000*weights[3] )/ pf['TSLA'].iloc[0]
    new_tsla = num_tsla.astype(int) * pf['TSLA']
    num_nvda= (100000*weights[4] )/ pf['NVDA'].iloc[0]
    new_nvda = num_nvda.astype(int) * pf['NVDA']  
    total = pd.DataFrame()
    total=new_goog + new_msft + new_aapl +new_nvda+new_tsla
    return total
   
    
plt.plot(analyze_better_portfolio(
    ticker=['AAPL', 'MSFT', 'GOOG', 'TSLA', 'NVDA'], 
    weights=[0.2, 0.1, 0.1, 0.2, 0.4], 
    start_date="2023-01-01", 
    end_date="2023-12-31"
) ,linestyle ='--',label="With new portfolio")
plt.plot(analyze_better_portfolio(
    ticker=['AAPL', 'MSFT', 'GOOG', 'TSLA', 'NVDA'], 
    weights=[0.3, 0.2, 0.2, 0.2, 0.1], 
    start_date="2023-01-01", 
    end_date="2023-12-31"
) , label="With old portfolio")
plt.xlabel('Time')
plt.ylabel('Portfolio Value')
plt.title("Portfolio's Performance")
plt.legend()
plt.show()

    
```
